## The Best Neighborhood

```python
import pandas as pd

chip = pd.read_csv("https://data.wprdc.org/dataset/46fb3ca4-e844-4b42-b034-e87291d34699/resource/06664b02-c673-49d5-8a70-d3cd1c18ac8d/download/pps_schoolsapr2019publish.csv")
ran = pd.read_csv("https://data.wprdc.org/dataset/27b6eef7-8825-4cc4-8b48-52c17ac80d37/resource/7b0660d1-c812-496d-ab0a-8560902e9e70/download/neighborhood_school_enrollment.csv")
chip.head()
ran.head()
#ran.plot.bar()
query_mask = ran['enrolled_students'] > 150
blah = ran[query_mask]
blah.plot.bar(x = "neighborhood" , rot = 90)
```

I think the best neighborhood to enroll a kid is in Brookline, PA. I believe that this will be the best school for a kid to enroll because it has the most enrolled students in this particular neighborhood. I believe that since this neighborhood is surrounded by more kids, new enrolled kids will be able to succeed together because their neighborhood school has the most enrolled students. Brookline, PA has almost 350+ students enrolled into the neighborhood school. Compared to the other neighborhoods, there is no other neighborhood who has as much students enrolled into the school. It will allow several kids to develop a friendship within the neighborhood, and also kids that are around their age.

```python
import pandas as pd
from matplotlib import pyplot as plt

sorted = ran.sort_values(by = ['enrolled_students'])

sorted.head(5)
lowest = sorted.head(5)

sorted.tail(5)
highest = sorted.tail(5)

fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize = (30,10))

#plt.subplot(1,3,1)
#blah.plot.bar(x = "neighborhood" , rot = 90)
ax1.bar(blah["neighborhood"],blah["enrolled_students"], color = ["black", "pink","red"])
ax1.tick_params('x', labelrotation = 90)
ax1.title.set_text('Neighborhoods')

#xticks(rotation = 90)
#plt.plot(blah)

#plt.subplot(1,3,2)
ax2.bar(lowest["neighborhood"],lowest["enrolled_students"])
ax2.tick_params('x', labelrotation = 90)
ax2.title.set_text('LOWEST 5')

#plt.subplot(1,3,3)
#plt.plot(highest)
ax3.bar(highest["neighborhood"],highest["enrolled_students"])
ax3.tick_params('x', labelrotation = 90)
ax3.title.set_text('TOP 5')

#ax1.figure(figsize = (190,190))
```
